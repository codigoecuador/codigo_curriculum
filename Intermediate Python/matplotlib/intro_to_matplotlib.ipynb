Matplotlib is a drawing library. It is very useful for data science when we want to display our results in a simple chart. In this section, we will examine the `matplotlib.pyplot` module in the library.

```python
import matplotlib.pyplot as plt #import the pyplot module from the matplotlib library
import numpy as np #import the numpy library
```

#### Example: Sine function

```python
#The sine is a function from mathematics. If you've taken a class in mathematics, you 
#know that a triangle has three sides, and three angles. The sine function in a right
#triangle is defined as the length of the opposite side divided by the
#length of the hypotenuse

# Let's calculate the x and y coordinates of each point in the sine curve
x = np.arange(0, 3 * np.pi, 0.1) # Values ranging from 0 to 3 * pi 0.1
y = np.sin(x) # Let's calculate sine values from x values

plt.plot(x, y) # plot the result
```

```python
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Sine and Cosine')
plt.legend(['sine', 'cosine'])
plt.savefig('sine_cosine.jpg') # save the file in your directory
```

### Subplots

We can show most the plot types visually with the `subplot` where they are a portion of a larger matplotlib plot.

```python
# Let's calculate the x and y coordinates for the sine and cosine
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Let's define Height: 2 and Width = 1
# So we can show up to 2 underlines.

# Define the first underline (Sines)
plt.subplot(2, 1, 1)
plt.plot(x, y_sin)
plt.title('Sines')

# Define the second underline (Cosines)
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cosines')

# Let's print the subplots on the screen
plt.show()
```
